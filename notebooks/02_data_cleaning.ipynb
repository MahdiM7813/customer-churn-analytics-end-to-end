# Data Cleaning

In this notebook, we transform the raw Telco Customer Churn dataset into a clean and analysis-ready format. We standardize column names, handle missing values, convert data types, normalize categorical values, and engineer new features such as tenure groups and revenue proxies. These steps ensure that subsequent analyses, visualizations, and dashboards are based on reliable, well-structured data.

The cleaning process follows best practices for data engineering and data science, preparing the dataset for exploratory analysis, SQL queries, and Power BI dashboards.

```python
# Load raw dataset for cleaning stage
import pandas as pd

df = pd.read_csv("E:/Customer Churn Analysis/data/raw/Telco-Customer-Churn.csv")

# Quick inspection
df.head()
```

```python
# Standardizing column names for consistency and SQL compatibility
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
# Identify missing values in dataset
df.isnull().sum()
```

```python
# Convert TotalCharges to numeric (fix hidden string values)
df["totalcharges"] = pd.to_numeric(df["totalcharges"], errors="coerce")
```

```python
# Handle missing values in TotalCharges
df["totalcharges"] = df["totalcharges"].fillna(df["totalcharges"].median())
```

```python
# Convert churn into binary variable for analysis and modeling
df["churn_flag"] = df["churn"].map({"Yes": 1, "No": 0})
```

```python
# Normalize categorical noise in service columns
replace_cols = [
    "multiplelines",
    "onlinesecurity",
    "onlinesecurity",
    "techsupport",
    "deviceprotection",
    "streamingtv",
    "streamingmovies"
]

for col in replace_cols:
    df[col] = df[col].replace("No internet service", "No")
    df[col] = df[col].replace("No phone service", "No")
```

```python
# Create meaningful customer lifecycle segments
df["tenure_group"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 24, 48, 72],
    labels=["0-1 year", "1-2 years", "2-4 years", "4+ years"]
)
```

```python
# Estimate total customer value contribution
df["total_revenue"] = df["monthlycharges"] * df["tenure"]
```

```python
# Save cleaned dataset for EDA, SQL and Power BI
df.to_csv("E:/Customer Churn Analysis/data/processed/telco_churn_cleaned.csv", index=False)
```
